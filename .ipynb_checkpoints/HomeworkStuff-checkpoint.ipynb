```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import tree
from sklearn.metrics import confusion_matrix
from sklearn import naive_bayes
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn import decomposition
from sklearn import preprocessing
import seaborn as sns
%matplotlib inline
```

```python
train = pd.read_csv('digit-recognizer/train.csv')
test = pd.read_csv('digit-recognizer/test.csv')
```

```python
train.head(5)
```

```python
train_x = np.asarray(train.drop('label', axis=1))
train_y = np.asarray(train['label'])
```

```python
test_x = np.asarray(test)
```

```python
preprocess = Sequential()
# Relu necessary because input to multinomial naive bayes must be nonnegative
preprocess.add(Conv2D(2, kernel_size=(3,3), use_bias=False, activation='relu', input_shape=(28,28,1)))
preprocess.add(MaxPooling2D(pool_size=(2,2)))
preprocess.add(Flatten())
preprocess.compile(loss='sparse_categorical_crossentropy')
preprocess.summary()
```

```python
weight_matrix = [np.asarray([[[[1,1]],[[2,0]],[[1,-1]]], [[[0,2]],[[0,0]],[[0,-2]]], [[[-1,1]],[[-2,0]],[[-1,-1]]]])]
preprocess.set_weights(weight_matrix)
```

```python
prepro_train = preprocess.predict(train_x.reshape((42000, 28, 28, 1)))
prepro_test = preprocess.predict(test_x.reshape((28000, 28, 28, 1)))
```

```python
# Another kind of preprocessing: PCA
# Unsupervised, so I can combine train and test (without labels) and use them together to find the principal components
```

```python
combined = np.concatenate((train_x, test_x))
pca = decomposition.PCA(whiten=False) #n_components (full)
combined_pca = pca.fit_transform(combined)
```

```python
plt.plot(pca.explained_variance_ratio_)
```

```python
trans_train_x = combined_pca[0:42000,:]
trans_test_x = combined_pca[42000:,:]
```

```python
tempdf = pd.DataFrame(trans_train_x)
tempdf['label'] = train_y
```

```python
sns.pairplot(x_vars=[0], y_vars=[1], data=tempdf, hue="label", size=8, plot_kws=dict(alpha=0.5))
```

```python
stand_train_x = preprocessing.scale(train_x)
#stand_test_x = preprocessing.scale(test_x)
stand_trans_train_x = preprocessing.scale(trans_train_x)
#stand_trans_test_x = preprocessing.scale(trans_test_x)
min_max_scaler = preprocessing.MinMaxScaler()
norm_train_x = min_max_scaler.fit_transform(train_x)
#norm_test_x = min_max_scaler.fit_transform(test_x)
norm_trans_train_x = min_max_scaler.fit_transform(trans_train_x)
#norm_trans_test_x = min_max_scaler.fit_transform(trans_test_x)
```

```python
# Initial decision tree model 
```

```python
np.random.seed(1)
traindata_x, testdata_x, traindata_y, testdata_y = train_test_split(train_x, train_y)
```

```python
np.random.seed(1)
standtraindata_x, standtestdata_x, standtraindata_y, standtestdata_y = train_test_split(stand_train_x, train_y)
np.random.seed(1)
normtraindata_x, normtestdata_x, normtraindata_y, normtestdata_y = train_test_split(norm_train_x, train_y)
```

```python
# Base
dtc = tree.DecisionTreeClassifier()
dtc.fit(traindata_x, traindata_y)
dtc_preds = dtc.predict(testdata_x)
dtc_cm = confusion_matrix(testdata_y, dtc_preds)
print(dtc_cm)
print(np.sum(np.diag(dtc_cm))/len(dtc_preds))
```

```python
# Standardized
dtc = tree.DecisionTreeClassifier()
dtc.fit(standtraindata_x, standtraindata_y)
dtc_preds = dtc.predict(standtestdata_x)
dtc_cm = confusion_matrix(standtestdata_y, dtc_preds)
print(dtc_cm)
print(np.sum(np.diag(dtc_cm))/len(dtc_preds))
```

```python
# Normalized
dtc = tree.DecisionTreeClassifier()
dtc.fit(normtraindata_x, normtraindata_y)
dtc_preds = dtc.predict(normtestdata_x)
dtc_cm = confusion_matrix(normtestdata_y, dtc_preds)
print(dtc_cm)
print(np.sum(np.diag(dtc_cm))/len(dtc_preds))
```

```python
# Tuned base (standardized has same accuracy, normalized has less...)
dtc2 = tree.DecisionTreeClassifier(criterion='entropy', min_impurity_decrease=0.0003)
dtc2.fit(traindata_x, traindata_y)
dtc2_preds = dtc2.predict(testdata_x)
dtc2_cm = confusion_matrix(testdata_y, dtc2_preds)
print(dtc2_cm)
print(np.sum(np.diag(dtc2_cm))/len(dtc2_preds))
```

```python
np.random.seed(1)
transtraindata_x, transtestdata_x, transtraindata_y, transtestdata_y = train_test_split(trans_train_x, train_y)
np.random.seed(1)
standtranstraindata_x, standtranstestdata_x, standtranstraindata_y, standtranstestdata_y = train_test_split(stand_trans_train_x, train_y)
np.random.seed(1)
normtranstraindata_x, normtranstestdata_x, normtranstraindata_y, normtranstestdata_y = train_test_split(norm_trans_train_x, train_y)
```

```python
# Standardized
rf = RandomForestClassifier(n_estimators=500)
rf.fit(standtraindata_x, standtraindata_y)
rf_preds = rf.predict(standtestdata_x)
rf_cm = confusion_matrix(standtestdata_y, rf_preds)
print(rf_cm)
print(np.sum(np.diag(rf_cm))/len(rf_preds))
```

```python
# Tuned standardized
rf = RandomForestClassifier(n_estimators=500, criterion='gini', min_impurity_decrease=0.00001)
rf.fit(standtraindata_x, standtraindata_y)
rf_preds = rf.predict(standtestdata_x)
rf_cm = confusion_matrix(standtestdata_y, rf_preds)
print(rf_cm)
print(np.sum(np.diag(rf_cm))/len(rf_preds))
```

```python
# Now that the best model has been found, need to run cross-validation on it to be sure
dtc = tree.DecisionTreeClassifier(criterion='entropy', min_impurity_decrease=0.0003)
mnb = naive_bayes.MultinomialNB(alpha=0.0)
dt_scores = cross_val_score(dtc, train_x, train_y, cv=10)
nb_scores = cross_val_score(mnb, train_x, train_y, cv=10)
```

```python
dtc_base = tree.DecisionTreeClassifier()
dt_base_scores = cross_val_score(dtc_base, train_x, train_y, cv=10)
```

```python
print("Base decision tree accuracy: %0.4f (+/- %0.4f)" % (100*dt_base_scores.mean(), 100*dt_base_scores.std() * 2))
print("Tuned decision tree accuracy: %0.4f (+/- %0.4f)" % (100*dt_scores.mean(), 100*dt_scores.std() * 2))
print("Naive Bayes accuracy: %0.4f (+/- %0.4f)" % (100*nb_scores.mean(), 100*nb_scores.std() * 2))
```
